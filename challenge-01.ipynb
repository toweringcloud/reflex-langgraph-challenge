```python
import openai
import os
from dotenv import load_dotenv

load_dotenv()
client = openai.OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

PROMPT = """
다음과 같이 영화 정보를 제공하는 API 서비스가 있습니다.
1) get_popular_movies() -> /movies API에서 인기 영화를 가져옵니다.
2) get_movie_details(id) -> /movies/:id API에서 영화 정보를 가져옵니다.
3) get_movie_credits(id) -> /movies/:id/credits API에서 출연진 및 제작진을 가져옵니다.

영화 정보 API의 기본 URL은 다음 호스트 정보를 참고해 주세요.
https://nomad-movies.nomadcoders.workers.dev

id는 영화의 고유한 식별자입니다. 인기 영화 목록에서 각 영화에 대한 ID를 확인할 수 있습니다.

영화 정보를 요청할 때, 필요한 정보의 유형에 따라 API 조회 결과를 한국어로 응답해줘!
- 목록 데이터의 경우, Max 5건의 데이터를 필터링 후, 영화 제목과 ID만 알려줘! 
- 상세 데이터의 경우, 영화 제목, 개봉 날짜, 장르, 줄거리 요약을 알려줘! 
- 출연진 데이터의 경우, 주요 배우들의 이름과 역할을 알려줘!
"""
```

```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": f"{PROMPT}\n지금 인기있는 영화가 무엇인지 알려줘!"}],
)
response.choices[0].message.content
```

```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": f"{PROMPT}\nmovie ID 550에 해당하는 영화가 무엇인지 알려줘!"}],
)
response.choices[0].message.content
```

```python
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": f"{PROMPT}\nmovie ID 550에 해당하는 영화에 누가 출연하는지 알려줘!"}],
)
response.choices[0].message.content
```
